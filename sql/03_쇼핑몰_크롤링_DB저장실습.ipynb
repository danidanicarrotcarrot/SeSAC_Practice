## 1. Crawling
- import

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
```

- webdriver 실행

```python
# Chrome WebDriver 경로 설정
chrome_driver_path = "./chromedriver.exe" # 프로젝트와 같은 경로
service = Service(chrome_driver_path)

# WebDriver 실행
options = webdriver.ChromeOptions()
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(service=service, options=options)
```

- 사이트 크롤링

```python
# 사이트 url
url = "https://jolse.com/category/toners-mists/1019?page="

# 데이터 저장 리스트 초기화
titles, origin_prices, discount_prices, links, imgs = [], [], [], [], []

# 사이트 열기
for i in range(1, 14): # 한 페이지 당 40개 * 10 페이지
    driver.get(url+str(i))
    time.sleep(3)

    # 각 페이지 1~40개 데이터 크롤링
    for j in range(1, 41):
        try:
            title = driver.find_element(By.XPATH, f'//li[{j}]/div[4]/strong/a/span[2]').text
            origin_price = driver.find_element(By.XPATH, f'//li[{j}]/div[4]/ul/li[1]/span[1]').text
            discount_price = driver.find_element(By.XPATH, f'//li[{j}]/div[4]/ul/li[2]/span[1]').text
            link = driver.find_element(By.XPATH, f'//li[{j}]/div[4]/strong/a').get_attribute('href')
            img = driver.find_element(By.XPATH, f'//li[{j}]/div[1]/div/a/img').get_attribute('src')
            
            titles.append(title)
            origin_prices.append(origin_price)
            discount_prices.append(discount_price)
            links.append(link)
            imgs.append(img)
            
        except Exception as e:
            print(f"{i}페이지 {j}번째 데이터를 가져오는 중 오류 발생: {e}")

# WebDriver 종료
driver.quit()
```

- 크롤링 데이터 insert

```python
for i in range(len(titles)):
    ti = titles[i]
    op = origin_prices[i]
    dp = discount_prices[i]
    li = links[i]
    im = imgs[i]
    
    sql = "INSERT INTO product (title, ori_price, discount_price, link, img) values(%s, %s, %s, %s, %s)"
    cursor.execute(sql, (ti, op, dp, li, im))
    db.commit()
    
```
