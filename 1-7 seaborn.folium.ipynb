```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings(action='ignore')

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='gulim')
```

```python
x = np.linspace(1, 14, 100)
x
```

```python
y1 = np.sin(x)
y2 = 2*np.sin(x)
y3 = 3*np.sin(x)
y4 = 4*np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
import seaborn as sns
```

```python
x = np.linspace(1, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x)
y3 = 3*np.sin(x)
y4 = 4*np.sin(x)

sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
sns.set_style('dark')
plt.figure(figsize=(10, 6))

plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))

plt.plot(x, y1, x, y2, x, y3, x, y4)
sns.despine(right=False, left=False)
plt.show()
```

```python
x = np.linspace(1, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x)
y3 = 3*np.sin(x)
y4 = 4*np.sin(x)

sns.set_context('talk')
# plt.figure(figsize=(10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
x = np.linspace(1, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x)
y3 = 3*np.sin(x)
y4 = 4*np.sin(x)

sns.set_context('talk', font_scale=.5)
plt.figure(figsize=(10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

# sns의 기본 데이터 셋 : tips 활용

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.histplot(x=tips['total_bill'])
```

```python
sns.kdeplot(x=tips['total_bill'])
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))
sns.boxplot(x=tips['total_bill'])
sns.despine()
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))
sns.boxplot(x=tips['day'], y=tips['total_bill'])
sns.despine(offset=10)
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', data=tips, palette='Set2')
sns.despine(offset=10)
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='tip', hue='smoker', data=tips, palette='Set3')
sns.despine(offset=10)
plt.show()
```

```python
# 데이터가 모여 잇는 정도를 보여주는 plot: seaborn.swarmplot()

plt.figure(figsize=(10, 6))
sns.swarmplot(x='day', y='total_bill', data=tips, palette='husl')
plt.show()
```

```python
# 관련 데이터 가져오기
state_unemployment = "./data/02. folium_US_Unemployment_Oct2012.csv"
state_unemp_data = pd.read_csv(state_unemployment)
state_unemp_data.head()
```

```python
state_geo = "./data/02. folium_us-states.json"
state_geo
```

```python
map = folium.Map(location=[340, -98], zoom_start=3)

# map
folium.Choropleth(geo_data=state_geo, data=state_unemp_data,
                  columns=['State', 'Unemployment'],
                  key_on = 'feature.id',
                  fill_color='YlGn',
                  legend_name='Unemployment Rate(%)').add_to(map)

folium.LayerControl().add_to(map)
map
```
